```python
# Github Link : https://github.com/Amanjakhetiya/OSS_LAB_5

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1
X=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(X),np.sin(X)

plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
# 2
plt.figure(figsize=(8,6),dpi=80)
plt.subplot(1,1,1)

plt.plot(X,C,color='blue',linewidth=1.0,linestyle='-')
plt.plot(X,S,color='green',linewidth=1.0,linestyle='-')

plt.xlim(-4.0,4.0)
plt.xticks(np.linspace(-4,4,9,endpoint=True))

plt.ylim(-1.0,1.0)
plt.yticks(np.linspace(-1,1,5,endpoint=True))

plt.savefig("exercise_2.png",dpi=72)
```

```python
# 3

#Adding legends
plt.plot(X,C,color='blue',linewidth=2.5,linestyle='-',label='cosine')
plt.plot(X,S,color='green',linewidth=2.5,linestyle='-',label='sine')

plt.legend(loc='upper left')
```

```python
# 4

# Regular Plots
n=256
X=np.linspace(-np.pi,np.pi,n,endpoint=True)
Y=np.sin(2*X)

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
```

```python
# 5

#Scatter plot

n=1024

X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)

plt.scatter(X,Y)
```

```python
# 6

#Bar Chart

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='black')
plt.show()
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='black')
plt.show()
```

```python
X = np.linspace(-np.pi,np.pi,7,endpoint=True)

SIN=np.sin(X)
COS=np.cos(X)
TAN=np.tan(X)
COT=np.arctan(TAN)
SEC=np.arccos(COS)
COSEC=np.arcsin(SIN)

plt.figure(figsize=(8,5),dpi=80)

plt.plot(X,TAN,color='blue',linewidth=2.5,linestyle='-',label='tan')
plt.plot(X,COT,color='green',linewidth=2.5,linestyle='--',label='cot')
plt.plot(X,SEC,color='red',linewidth=2.5,linestyle='-.',label='sec')
plt.plot(X,COSEC,color='black',linewidth=2.5,linestyle=':',label='cosec')

plt.legend(loc='upper center')
```
